```python
# Import
from ultralytics import YOLO
```

```python
# Load Trained Model
model = YOLO("../runs/detect/train3/weights/best.pt")
```

```python
# Show Predictions
import matplotlib.pyplot as plt
import cv2
import os

result_dir = "../runs/detect/predict"

images = os.listdir(result_dir)

for img_name in images[:5]:
    img = cv2.imread(os.path.join(result_dir, img_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img)
    plt.axis("off")
    plt.show()
```
